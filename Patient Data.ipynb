```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import math
```

```python
patient= pd.read_csv("COVID_19.csv")
patient.head(5)
```

```python
print('no of patient in original data:' +str(len(patient.index)))
```

## Data Wrangling

```python
patient.isnull().sum()
```

```python
clean_num = {"gender": {"Male":1, "Female":0},
             "Label":  {"Positive":1, "Negative":0}}
patient.replace(clean_num, inplace=True)
patient.head(5)
```

```python
patient.drop(['Region2','nationality'],axis=1,inplace=True)
```

```python
patient.head(5)
```

```python
label_encoder=LabelEncoder()
patient['Travel_hist']=  label_encoder.fit_transform(patient['Travel_hist'])  
patient['detected_state']=  label_encoder.fit_transform(patient['detected_state'])
#patient['Travel_hist'].unique() 
patient.head(5)
```

```python
Symp= pd.get_dummies(patient['Symptom'])
Symp.head(5)
```

```python
Dis= pd.get_dummies(patient['Disease_hist'])
Dis.head(5)
```

```python
patient= pd.concat([patient,Symp,Dis],axis=1)
patient.head(5)
```

```python
patient.drop(['Symptom','Disease_hist'],axis=1,inplace=True)
patient.head(10)
```
